```python
import pandas as pd

# Load the CSV files from hammer-5-csv folder
product_df = pd.read_csv('hammer-5-csv/hammer-4-product.csv')
raw_df = pd.read_csv('hammer-5-csv/hammer-4-raw.csv')

# Display the first few rows to verify
print("Product data:")
print(product_df.head(10))
print("\nRaw data:")
print(raw_df.head(10))
```

```python
import pandas as pd

# 1. Ensure nowtime is in datetime format (crucial for accurate comparison)
raw_df['nowtime'] = pd.to_datetime(raw_df['nowtime'])

# 2. Find the index of the latest record for each product_id
# This avoids sorting the whole dataframe
latest_indices = raw_df.groupby('product_id')['nowtime'].idxmax()

# 3. Use those indices to pull the full rows from raw_df
latest_raw = raw_df.loc[latest_indices]

# 4. Merge with product_df
# We specify only the columns we need from latest_raw to save memory
final_df = product_df.merge(
    latest_raw[['product_id', 'current_price', 'old_price', 'price_per_unit', 'other']], 
    left_on='id', 
    right_on='product_id', 
    how='inner'
)

# 5. Convert to list of dictionaries
result = final_df.to_dict('records')

print(f"Total records: {len(result)}")
```

```python
# save the result to a json file
import json
with open('latest_grocery_data.json', 'w') as f:
    json.dump(result, f, indent=4)
```

```python
# load the JSON file to verify
with open('latest_grocery_data.json', 'r') as f:
    data = json.load(f)

print(f"Loaded {len(data)} records from JSON file.")

print(data[:10])
```
